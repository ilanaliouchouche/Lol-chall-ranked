```python
tier_cols = ['100_TOP_tier', '100_JUNGLE_tier', '100_MIDDLE_tier', '100_BOTTOM_tier', '100_UTILITY_tier',
             '200_TOP_tier', '200_JUNGLE_tier', '200_MIDDLE_tier', '200_BOTTOM_tier', '200_UTILITY_tier']
df[tier_cols] = df[tier_cols].apply(lambda x: x.str.slice(0, 3))

print(df['100_TOP_tier'].unique())

tier_df = df[tier_cols]
tier_df
```

```python
mode = tier_df.drop('100_TOP_tier', axis=1).mode().iloc[0].mode()
df['100_TOP_tier'].fillna(mode, inplace=True)
```
